#The example below shows you how to display simple geometry using vtk PolyData 

```python

# this Example shows you how to use vtkPolyDataMapper to display simple geometry 
from vtk import *
 
# You can read a vtp source file as the input 
file_name = "disk_out_ref_surface.vtp"
reader = vtkXMLPolyDataReader()
reader.SetFileName(file_name)
reader.Update() # Needed because of GetScalarRange

# -------   or you can read a STL file as the input 
stl_name = "42400-IDGH.stl" 
stl_reader = vtkSTLReader()
stl_reader.SetFileName(stl_name)
stl_reader.Update()

# -------   Or you can create a cylinder using vtk cylinder source 
cylinder = vtkCylinderSource()
cylinder.SetResolution(80) 

# --------  Or you can read a vtk structured file, but you will need a vtkStructuredGridGeometryFilter 
str_name = "multicomb_1.vts" 
str_l_name = "SampleStructGrid.vtk" 

str_reader = vtkXMLStructuredGridReader()   # vtk XML structured grid data 
str_reader2 = vtkStructuredGridReader()     # or vtk Legacy structured grid data 

str_reader.SetFileName(str_name) 
str_reader2.SetFileName(str_l_name) 

str_reader2.Update() 
str_reader.Update()

# --------  Or you can read a vtk structured file, but you will need a vtkStructuredGridGeometryFilter 


geomFilter = vtkStructuredGridGeometryFilter()
#geomFilter.SetInputConnection(str_reader.GetOutputPort())
geomFilter.SetInputConnection(str_reader2.GetOutputPort())  
geomFilter.Update() 

#--------------------------------------------------------
# comment and uncomment the code below depending on how you want to set the mapper input 
# Create the mapper that corresponds the objects of the vtk file
# into graphics elements
#mapper = vtkPolyDataMapper()

#Set the object read from the input file to the mapper 
#mapper.SetInputConnection(reader.GetOutputPort())

# or you can set the stl source as the input to the mapper 
#mapper.SetInputConnection(stl_reader.GetOutputPort())

# or you can set the cylinder source geometry as the input to the mapper 
#mapper.SetInputConnection(cylinder.GetOutputPort())

# or you can set the structured grid geometry as the input to the mapper 
mapper.SetInputConnection(geomFilter.GetOutputPort())
 
# Create an Actor
actor = vtkActor()
actor.SetMapper(mapper)
 
# Create a Renderer
renderer = vtkRenderer()
renderer.AddActor(actor)
renderer.SetBackground(1, 0, 1) # Set background to white
 
# Create a RendererWindow
renderer_window = vtkRenderWindow()
renderer_window.AddRenderer(renderer)
 
# Create a RendererWindowInteractor and display the vtk object 
interactor = vtkRenderWindowInteractor()
interactor.SetRenderWindow(renderer_window)
interactor.Initialize()
interactor.Start()
```

# An example how to read and display a rectilinear mesh

```python
from vtk import *
 
# The source file
file_name = "cth.vtr"
 
# Read the source file.
reader = vtkXMLRectilinearGridReader()
reader.SetFileName(file_name)
reader.Update() 

# Create the mapper that corresponds the objects of the vtk file
# into graphics elements
mapper = vtkDataSetMapper()
mapper.SetInputConnection(reader.GetOutputPort())
 
# Create the Actor
actor = vtkActor()
actor.SetMapper(mapper)
actor.GetProperty().SetRepresentationToWireframe()
 
# Create the Renderer
renderer = vtkRenderer()
renderer.AddActor(actor)

renderer.SetBackground(.3, .6, .3) # Set background to white
 
# Create the RendererWindow
renderer_window = vtkRenderWindow()
renderer_window.AddRenderer(renderer)

renderer_window.Render()
 
# Create the RendererWindowInteractor and display the vtk_file
interactor = vtkRenderWindowInteractor()
interactor.SetRenderWindow(renderer_window)
interactor.Initialize()
interactor.Start()
```

#Example to show how to read and display unstructured grid data 

```python
from vtk import *
 
# The source file
file_name = "uGridEx.vtk"
 
# Read the source file.
reader = vtkUnstructuredGridReader()
reader.SetFileName(file_name)
reader.Update()
 
# Create the mapper that corresponds the objects of the vtk file
# into graphics elements
mapper = vtkDataSetMapper()
mapper.SetInputConnection(reader.GetOutputPort())
 
# Create the Actor
actor = vtkActor()
actor.SetMapper(mapper)
 
# Create the Renderer
renderer = vtkRenderer()
renderer.AddActor(actor)
renderer.SetBackground(.3, .3, .3) # Set background to white
 
# Create the RendererWindow
renderer_window = vtkRenderWindow()
renderer_window.AddRenderer(renderer)
 
# Create the RendererWindowInteractor and display the vtk_file
interactor = vtkRenderWindowInteractor()
interactor.SetRenderWindow(renderer_window)
interactor.Initialize()
interactor.Start()
```

# Below is a python script that will take an input and tell you what type the data set is 

```python
import sys 
from vtk import *

vtk_cell_type = {1:'vertex', 2:'poly_vertex', 3:'line', 4:'poly_line', 5:'triangle',
                 6:'triangle_strip', 7:'polygon', 8:'pixel', 9:'quad', 10:'tetra',
                 11:'voxel', 12:'hexahedron'}

# below is used if you run the python program in command line 
#if (len(sys.argv)!=2):
#    print("usage: vtk_generic_query.py <file>")
#    exit()
#
#file_name = sys.argv[1]
#print("file name = "+ file_name)

#for ipython noteook, use input()
#for example, upon prompted, type "UGridEX.vtk"  -- don't forget about "" 
file_name = input()  


# Read the source file.
reader = vtkGenericDataObjectReader()
reader.SetFileName(file_name)
reader.Update() # Needed because of GetScalarRange

#------------------
if (reader.IsFilePolyData()):
    print("this is a vtkPolyData type")
    nop = reader.GetPolyDataOutput().GetNumberOfPoints()
    print 'number of points = ' + str(nop)
#------------------
elif (reader.IsFileStructuredGrid()):
    print("this is a vtkSturcturedGrid type")
    dims = reader.GetStructuredGridOutput().GetDimensions()
    nop = reader.GetStructuredGridOutput().GetNumberOfPoints()
    print dims, nop
#------------------    
elif (reader.IsFileUnstructuredGrid()):
    print("this is an vtkUnstructuredGrid tpye:")
    noc = reader.GetUnstructuredGridOutput().GetNumberOfCells()
    nop = reader.GetUnstructuredGridOutput().GetNumberOfPoints()
    cot = reader.GetUnstructuredGridOutput().GetCellType(0)
    cotk = reader.GetUnstructuredGridOutput().GetCellType(noc-1)
    print 'number of cells= '+str(noc)+' number of points = '+str(nop)
    print 'cell[0] type = '+ vtk_cell_type[cot] + 'cell['+str(noc-1)+'] type = ' + vtk_cell_type[cotk]
#------------------
elif (reader.IsFileRectilinearGrid()):
    print("this is an vtkRectilinearGrid tpye:")
    dims = reader.GetRectilinearGridOutput().GetDimensions()
    print dims
#------------------
elif (reader.IsFileStructuredPoints()):
    print("this is a vtkStructuredPoints tpye:")
    dims = reader.GetStructuredPointsOutput().GetDimensions()
    print dims

```
